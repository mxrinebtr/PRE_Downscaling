```python
os.chdir('/work/FAC/FGSE/IDYST/tbeucler/downscaling/mberthier/repos/Downscaling_CM/data')
gev_params=xr.open_dataset('GEV_params_stationnary.nc')
```

```python
from sklearn.linear_model import LinearRegression

data=gev_params
lon = data['rlon'].values
lat = data['rlat'].values

# Concaténez les variables d'entrée en une seule matrice

lon_mesh, lat_mesh = np.meshgrid(lon, lat)
X = np.column_stack((lon_mesh.flatten(), lat_mesh.flatten()))

shape = data['shape'].values
location = data['location'].values
scale = data['scale'].values

y = np.column_stack((shape.flatten(), location.flatten(), scale.flatten()))

model = LinearRegression()

# Ajustez le modèle aux données d'entraînement
model.fit(X, y)

coefficients = model.coef_  # Coefficients de chaque variable d'entrée
intercept = model.intercept_  # Terme d'interception

print(coefficients)
print(intercept)
print(model)
#print(predictions_xr)
```
